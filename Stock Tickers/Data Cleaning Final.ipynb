````python

import glob

path = r"C:\Users\night\RRL-Stock-Trading\Stock Tickers\AAPL"
name = 'AAPL' # change the ticker name
files = glob.glob(path + "\*.csv")

for f in files:
  print(f)
````

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame()
for yr in files:
    df = pd.concat([pd.read_csv(yr, parse_dates=True, index_col=1), df])
df.sort_index(inplace=True)
df.drop(['symbol'], axis=1, inplace=True)
df.head()
````

````python

df = df[["close", "high", "low", "open", "volume"]]
df.head(15)
````

````python
## Verify Data has loaded 
print(f"No. of data points: {len(df)}")
print(f"Start of raw data: {df.index[0]}")
print(f"End of raw data: {df.index[-1]}")
````

# Data Preprocessing 

````python
start = '2017-01' ## YYYY-MM
end = '2021-12' ## YYYY-MM

df = df.loc[start:end].copy()

print("After filtering:")
print(f"No. of data points: {len(df)}")
print(f"Start of raw data: {df.index[0]}")
print(f"End of raw data: {df.index[-1]}")
````

````python
# 1hr df2
df2 = pd.DataFrame()

df2['close'] = df.close.resample('60min').last()
df2['high'] = df.high.resample('60min').max()
df2['low'] = df.low.resample('60min').min()
df2['open'] = df.open.resample('60min').first()
df2['volume'] = df.volume.resample('60min').sum()
````

````python

ticker = df2
ticker.drop(['close', 'low', 'open'], axis=1, inplace=True)
ticker.dropna(inplace=True) 
ticker.head()
````
